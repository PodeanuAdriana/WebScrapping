```python
import pandas as pd
import math
import numpy as np

df=pd.read_csv('scrapping_booking.csv')
pret_noapte=df['Pret_noapte']/1000
nr_nopti=df['Nr_nopti']
pret_total=(df['Pret_noapte']*df['Nr_nopti'])/1000
# pret
pret_total=(pret_total.replace('.',','))
# nr_nopti=df['Nr_nopti']
scor=df['Scor']
# print(df.head())
print(pret_total)
print(pret_noapte)
lung=nr_nopti.count()
print(lung)
suma_preturi_tot=pret_total.sum()
medie_preturi_tot=suma_preturi_tot/lung
print(suma_preturi_tot)
print(medie_preturi_tot)
suma_scoruri=scor.sum()
medie_scoruri=suma_scoruri/lung

print(medie_scoruri)

preturi_centrate=pret_total-medie_preturi_tot
preturi_centrate_2=(preturi_centrate**2).sum()
print(preturi_centrate)
scoruri_centrate=scor-medie_scoruri
scoruri_centrate_2=(scoruri_centrate**2).sum()
print(scoruri_centrate)
print(scoruri_centrate_2)
dependenta=preturi_centrate*scoruri_centrate

suma_depend=dependenta.sum()
# print(dependenta)
print(suma_depend)
r=suma_depend/np.sqrt(preturi_centrate_2 * scoruri_centrate_2)
print(r)
nr_stele=df['Nr_stele']
# print(nr_stele.head())
corelatie = pret_total.corr(nr_stele)

print(f"Coeficientul de corelație între preț și numărul de stele: {corelatie}")
```

```python
chunk_size = 219 

for i in range(0, len(df), chunk_size):
    chunk = df.iloc[i:i+chunk_size]
    
    Creăm două coloane noi cu zilele săptămânii
    chunk['Check-in Day'] = chunk['Check-in'].dt.day_name()
    chunk['Check-out Day'] = chunk['Check-out'].dt.day_name()

    # Filtrăm zilele lucrătoare (Monday - Friday)
    df_week_days = (chunk['Check-in Day'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']) & 
                    chunk['Check-out Day'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']))
    
    df_pret_weekdays = chunk[df_week_days]['Pret_noapte']

    # Filtrăm zilele de weekend (Saturday & Sunday)
    df_weekend_days = (chunk['Check-in'].dt.day_name().isin(['Saturday', 'Sunday'])) | \
                      (chunk['Check-out'].dt.day_name().isin(['Saturday', 'Sunday']))
    
    df_pret_weekend = chunk[df_weekend_days]['Pret_noapte']

    # Calculăm media prețurilor
    medie_weekend_preturi = df_pret_weekend.mean()
    medie_weekday_preturi = df_pret_weekdays.mean()
    
if medie_weekend_preturi > medie_weekday_preturi:
    print(f"Preturile sunt mai mari cu:{medie_weekend_preturi - medie_weekday_preturi:.2f} RON.")
elif medie_weekend_preturi < medie_weekday_preturi:
    print(f"Preturile sunt mai mici cu : {medie_weekday_preturi - medie_weekend_preturi:.2f} RON.")
else:
    print("Preturi egale") 
#     print(chunk[['Check-in Day', 'Check-out Day']])

print(f" Media pret weekend: {medie_weekend_preturi:.2f} RON")
print(f"Media pret zile lucratoare: {medie_weekday_preturi:.2f} RON")
```
